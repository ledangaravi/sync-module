# WEBENCH Power Designer Graphs

Create matplotlib graphs from WEBENCH simulation data in csv files.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
# font used in LaTex report
rc('font', **{'family': 'serif', 'serif': ['Computer Modern'], 'size': 16})
rc('text', usetex=True)
```

```python
figure_size = (10,5)
```

## TLV62130

```python
filepath = 'tlv62130.csv'
df = pd.read_csv(filepath)
```

```python
plt.figure(figsize=figure_size)
plt.plot(df['Output Current'], df['Vin=5.5V'], label='Vin=5.5V')
plt.plot(df['Output Current'], df['Vin=6.95V'], label='Vin=6.95V')
plt.plot(df['Output Current'], df['Vin=8.4V'], label='Vin=8.4V')
plt.legend()
plt.xscale('log')
plt.xlabel('Output current [A]')
plt.ylabel('Efficiency [\%]')
plt.savefig('tlv62130.pdf', bbox_inches='tight', dpi=300)
plt.show()
```
